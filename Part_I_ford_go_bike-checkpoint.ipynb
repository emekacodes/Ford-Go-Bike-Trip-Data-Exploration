## Preliminary Wrangling


```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
#Load the dataset
fordbike_df = pd.read_csv('201902-fordgobike-tripdata.csv')
```

```python
#View the head of the dataset
fordbike_df.head()
```

```python
#View the datsets' shape
fordbike_df.shape
```

```python
#View the datasets' info
fordbike_df.info()
```

```python
# Convert start_time and end_time to datetime objects
fordbike_df['start_time'] = pd.to_datetime(fordbike_df['start_time'])
fordbike_df['end_time'] = pd.to_datetime(fordbike_df['end_time'])

# Convert start_station_id and end_station_id to int datatypes
fordbike_df['start_station_id'] = fordbike_df['start_station_id'].astype('Int64')
fordbike_df['end_station_id'] = fordbike_df['end_station_id'].astype('Int64')

# Convert start_station_name, end_station_name, user_type, member_gender and bike_share_for_all_trip to category datatypes
fordbike_df['start_station_name'] = fordbike_df['start_station_name'].astype('category')
fordbike_df['end_station_name'] = fordbike_df['end_station_name'].astype('category')
fordbike_df['user_type'] = fordbike_df['user_type'].astype('category')
fordbike_df['member_gender'] = fordbike_df['member_gender'].astype('category')
fordbike_df['bike_share_for_all_trip'] = fordbike_df['bike_share_for_all_trip'].astype('category')
```

```python
#Viewing descriptice statistics for numerical variables/features
fordbike_df.describe()
```

```python
#Drop all null rows to enable visualizations plot properly
fordbike_df.drop(fordbike_df[fordbike_df.start_station_id.isnull()].index, inplace = True)
fordbike_df.drop(fordbike_df[fordbike_df.start_station_name.isnull()].index, inplace = True)
fordbike_df.drop(fordbike_df[fordbike_df.end_station_id.isnull()].index, inplace = True)
fordbike_df.drop(fordbike_df[fordbike_df.end_station_name.isnull()].index, inplace = True)
fordbike_df.drop(fordbike_df[fordbike_df.member_birth_year.isnull()].index, inplace = True)
fordbike_df.drop(fordbike_df[fordbike_df.member_gender.isnull()].index, inplace = True)
```

## Univariate Exploration

Lets look at the distribution of the duration_sec plot

```python
#Standard plot of duration_sec

bin = np.arange(0, fordbike_df.duration_sec.max()+50, 50)
plt.hist(fordbike_df.duration_sec, bins = bin)
plt.xlabel('Duration (Seconds)');
```

The plot is long_tailed, lets use a log transformation for it

```python
#Log tranformed duration_sec histogram

bin = 10 ** np.arange(np.log10(fordbike_df.duration_sec.min()), np.log10(fordbike_df.duration_sec.max() + 0.05), 0.05)
plt.hist(fordbike_df.duration_sec, bins = bin)
plt.xscale('log')
plt.xlabel('Duration (Seconds)');
x_ticks = [100, 300, 1000, 3000, 10000, 30000, 100000]
xlabel = ['{}'.format(x) for x in x_ticks]
plt.xticks(x_ticks, xlabel);
```

The duration falls mostly  between 300 and 1000 seconds, from this we can clearly see the length of the majority of the rides. We can also see that the distribution is unimodal.

Although it is weird for someone to have a trip of up to 85,000 seconds, i'll ignore this; taking in the condition that the person possibly forgot to return it or could had an extended trip

Then we will plot the member_birth_year column to see the distribution

```python
#Standard plot of member_birth_year

bin = np.arange(1870, fordbike_df.member_birth_year.max()+2, 2)
plt.hist(fordbike_df.member_birth_year, bins = bin)
plt.xlabel('Birth Year');
```

We can see a leftly skewed plot, realistically someone born in 1880 will probably be dead; so we dig deeper

```python
#Log transformed plot of birth years

bin = 10 ** np.log10(np.arange(1920, fordbike_df.member_birth_year.max()+2, 2))
plt.hist(fordbike_df.member_birth_year, bins = bin)
plt.xscale('log')
x_scale= [1920, 1940, 1960, 1980, 2000]
xlabel = ['{}'.format(v) for v in x_scale]
plt.xticks(x_scale, xlabel)
plt.xlabel('Birth Year');
```

The plot is rightly skewed,but we can see better that a lot of the ages fall between 1980 and 2000 which shows that the average age of the bikers will be between 40 and 20
